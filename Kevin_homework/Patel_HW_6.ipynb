```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score, confusion_matrix
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
```

```python
data_df = pd.read_csv("credit_db.csv")
data_df.head()
```

```python
data_df['GENDER'].value_counts()
data_df['Car_Owner'].value_counts()
data_df['Propert_Owner'].value_counts()
data_df['Housing_type'].value_counts()
```

# 1

```python

data_df = pd.get_dummies(data_df, columns=['GENDER', 'Car_Owner', 'Propert_Owner', 
                                           'Type_Income', 'Marital_status', 'Housing_type', 'Type_Occupation', 'EDUCATION'])
data_df = data_df.fillna(data_df.mean())
data_df.head()
```

# 2
The model didn't peforme that well because this isn't a good approach for this dataset. Since we know the out can only be 1. approved for a credit card or 2. not approved.

```python
X = data_df.drop('Annual_income', axis=1)
y = data_df['Annual_income']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=35)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
y_pred
```

```python
accuracy_score = model.score(X_test, y_test)
accuracy_score
```

```python
r2_score(y_test, y_pred)
```

# 3
This model performed a little bit better then the linear regression.

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)

def opt_k(X, y, max_k=20, cv=5):
    k_values = list(range(1, max_k + 1))
    cv_scores = []

    for k in k_values:
        knn = KNeighborsRegressor(n_neighbors=k)
        scores = cross_val_score(knn, X, y, cv=cv, scoring='neg_mean_squared_error')
        cv_scores.append(-scores.mean())

    optimal_k = k_values[np.argmin(cv_scores)]
    return optimal_k
optimal_k = opt_k(X_train, y_train)
print(optimal_k)
```

```python
knn = KNeighborsRegressor(n_neighbors=optimal_k)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

r2 = r2_score(y_test, y_pred)
r2
```

```python
print(knn.score(X_test, y_test))
```
